```python
from bokeh.core.properties import value
from bokeh.io import show, output_notebook, reset_output
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
from bokeh.transform import dodge
```

```python
reset_output()
output_notebook()
```

```python
recursions = ['1', '5', '10', '15', '20', '25', '30', '35']
code_type = ['Python', 'C++', 'GPU']

data = {'recursions' : recursions,
        'Python'   : [0.1, 1, 4, 3, 2, 4, 4, 2],
        'C++'      : [5,   3, 3, 2, 4, 6, 4, 9],
        'GPU'      : [3,   2, 4, 4, 5, 3, 1, 2]}
```

```python
src = ColumnDataSource(data=data)

p = figure(x_range=recursions, y_range=(0, 10), plot_height=250, toolbar_location=None, tools="")
w=0.2

p.vbar(x=dodge('recursions', -0.25, range=p.x_range), top='Python', width=w, source=src, color="#c9d9d3", legend=value("Python"))

p.vbar(x=dodge('recursions',  0.0,  range=p.x_range), top='C++', width=w, source=src, color="#718dbf", legend=value("C++"))

p.vbar(x=dodge('recursions',  0.25, range=p.x_range), top='GPU', width=w, source=src, color="#e84d60", legend=value("GPU"))

p.x_range.range_padding = 0.1
p.xgrid.grid_line_color = None
p.legend.location = 'top_left'
p.legend.orientation = 'horizontal'

show(p)
```
